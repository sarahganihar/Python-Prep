# 1.Import headers and Data Cleansing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statistics import median 
import seaborn as sns; sns.set()
from mpl_toolkits.basemap import Basemap
%matplotlib inline
from itertools import chain
from datetime import datetime



df=pd.read_csv("h1b_kaggle.csv",index_col=0).dropna()




#Data cleansing
df['YEAR']= df['YEAR'].astype(int)

#Dataframe Analysis
print(df.info())
print(df.head())


```

# 2. Exploratory Analysis of H1B dataset

```python
#Exploratory Analysis
#Top 20 Employers with highest no.of H1B petitions
df.EMPLOYER_NAME.value_counts().nlargest(20).plot(kind='bar', figsize=(10,5))
plt.xlabel("Job Title")
plt.ylabel("No.Of Applications")
plt.show()
```

# 3. Analysing the Salary trend from 2011-2016

```python
#Salary/Wages trend from 2011-2016
plt.figure(figsize=(10,9))
#print(df.head())
df1= df.groupby(['YEAR'])[['PREVAILING_WAGE']].median()
plt.plot(df1, color='blue')
plt.ylabel("Wage")
plt.show()
```

# 4.Top 10 Job Titles with most no.of H1B Appliactions Submitted

```python
#Top 10Job Titles based of total no.of applications
plotData= df.JOB_TITLE.value_counts().nlargest(10).plot(kind='bar', color='green', figsize=(10,5))
plt.xlabel("Job Title")
plt.ylabel("No.Of Applications")
plt.savefig('topljobs.png',dpi=300, bbox_inches='tight')
plt.show()
```

# 5. Top 10 cities hiring the most

```python
plotData= df.WORKSITE.value_counts().nlargest(10).plot(kind='bar', color='orange', figsize=(10,5))
plt.xlabel("Worksite")
plt.ylabel("No.Of Applications")
plt.savefig('toploc.png',dpi=300, bbox_inches='tight')
plt.show()

```
